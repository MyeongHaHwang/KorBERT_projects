```python
import re
from collections import defaultdict

texts = list(map(lambda x : x[0].replace('\n', ' '), df.values))
letter_ = texts[0][1]
# patterns = re.compile('([A-Z])\w+([ ㄱ-ㅎㅏ-ㅣ가-힣])+')
patterns = re.compile('[ASC]l\d+l\d+l[ ㄱ-ㅎㅏ-ㅣ가-힣]*')
result = defaultdict(list)
for (ix, text) in enumerate(texts):
    for s in patterns.finditer(text):
        res = {}
        s = s.group().strip().split(letter_)
        res['Speaker'], res['Start'], res['End'], res['Text'] = s
        result[ix].append(res)
```
